#### How to read a correlation heatmap?

<small>

**1. Look at the color of each cell to see the strength and direction of the correlation.**<br>
**2. Darker colors indicate stronger correlations, while lighter colors indicate weaker correlations.**<br>
**3. Positive correlations (when one variable increases, the other variable tends to increase) are usually represented by warm colors, such as red or orange.**<br>
**4. Negative correlations (when one variable increases, the other variable tends to decrease) are usually represented by cool colors, such as blue or green.**

</small>
